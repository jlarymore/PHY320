```python
from pylab import *
%matplotlib inline
```

```python
Nx = 101
Ny = 201
x = linspace(0, 10, Nx)
y = linspace(0, 20, Ny)
z = zeros((Ny, Nx))
```

```python
for i in range(Nx):
    for j in range(Ny):
        z[j,i] = sin(x[i])*cos(y[j])        
```

```python
colorValues = linspace(-0.5, 0.5, 101)
contourf(x, y, z, colorValues)
colorbar()
grid(True)
show()
```
