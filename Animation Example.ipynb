**ANIMATION EXAMPLE:**

```python
from time import sleep # To set the animation frame rate 
from IPython.display import clear_output # To redraw
```

We are going to plot a **wave** traveling as a function of space and time, $f(x,t)$.  The wave is describe mathematically by the equation
$$f(x,t) = A \sin(k x - \omega t),$$
where $A$ is the amplitude, 
$x$ is the position, and
$t$ represents time.
$k$ and $\omega$ are parameters that relate to the wavelength and frequency of the wave.

```python
# Demonstrates creating an animation...
A = 1
k = 1
omega = 1
x = linspace(0, 10, 1000)
tArray = linspace(0, 5, 51)

for t in tArray:
    sleep(0.05) # Sets the maximum animation speed
    
    f = A * sin(k*x - omega*t)
    plot(x, f, linewidth = 2)
    ylim([-A-0.1, A+0.1])
    xlabel('x'); ylabel('f(x,t)')
    titleString = "Time (sec): " + str(t)
    title(titleString)
    clear_output(wait=True)
    grid(True); show()
```
